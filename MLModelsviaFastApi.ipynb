```python
#creating a copy of our dataset because we will apply many feature engineering steps on our dataset
f_copy =  f.copy()
f_copy
```

```python
# Feature 1: maqking a new column (bmi)
f_copy["bmi"] = f_copy["Weight"] / (f_copy["Height"] ** 2)
f_copy
```

```python
# Feature 2: Age_group
def age_group(Age):
  if Age < 25:
    return "young"
  elif Age < 45:
    return "adult"
  if Age < 60:
    return "middle_aged"
  return "senior"
```

```python
f_copy["Age_group"] = f_copy["Age"].apply(age_group)
f_copy
```

```python
#Feature 3: Lifestyle Risk
def lifestyle_risk(row):
  if row["smoker"] and row["bmi"] > 30:
    return "high"
  elif row["smoker"] and row["bmi"] > 27:
    return "medium"
  else:
    return "low"
```

```python
f_copy["Lifestyle_Risk"] = f_copy.apply(lifestyle_risk, axis=1)
f_copy
```

```python
#cities ko alag se process naa karke cities ko tiers me divide kar denge
tier_1_cities = {"Mumbai", "BLR", "Pondicherry", "Pune", "Chennai", "Kolkata"}

tier_2_cities = {"Jaipur", "Chandigarh", "Indore", "Patna", "Lucknow", "Coimbatore", "Ranchi", "Amristar", "Varanasi", "Guwahati", "Allahabad", "Aligarh", "Gaya"}


```

```python
#Feature 4:-> City tier
def city_tier(city):
    if city in tier_1_cities:
      return 1
    elif city in tier_2_cities:
      return 2
    else:
      return 3
```

```python
f_copy["City_tier"] =  f_copy["city"].apply(city_tier)
f_copy
```

```python
# ab upar yeh saare humaare naye features aa gaye hai, yeh sab new features banne ke baad, hum kuch puraane features ko delete kar denge like AGE kyuki uske badle age_group aa gayaa
#height and wei8 ki zaroorat nhi hai kyuki uske badle bmi aa gayaa
#smoker ki zaroorat nhi hai kyuki uske badle lifestyle_risk aa gayaa
#city ki zaroorat nhi hai kyuki uske badle city_tier aa gayaa

f_copy.drop(columns=["Age", "Weight", "Height", "smoker", "city"])[["income_lpa", "Occupation", "bmi", "Age_group", "Lifestyle_Risk", "City_tier", "income_lpa", "INSURANCE_PREMIUM_category"]]
```

***now we will train our Model using the above DATASET***

```python
#sabse pehle humne humara X and Y ko alg alag kar liiyaa, where Y is the TARGET column and X is our features
x = f_copy[["bmi", "Age_group", "Lifestyle_Risk", "City_tier", "income_lpa", "Occupation"]]
y = f_copy["INSURANCE_PREMIUM_category"]
```

```python
x
```

```python
y
```

```python
# Defined our CATEGORICAL and NUMERICAL features alag se, aise isiliye kiya yuki aaget on categorical features we will apply 1-hot encoding
categorical_features = ["Age_group", "Lifestyle_Risk", "City_tier", "Occupation"]
numerical_features = ["bmi", "income_lpa"]
```

```python
#created a column transformer and is columntransformer me saare categorical features ke upar 1-hot encoding lagaa diya aur
#saare numerical features ko as it is passthrough karaa diya, uske upar koi transformation nhi kiya
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(), categorical_features),
        ("num", "passthrough", numerical_features)
    ]
)
```

```python
#phir humne ek sk learn pipeline banai in 2 steps
#pehla, column transformer wala(preprocessor)
#2nd, hum us transformed data ke upar ek randomForestClassifier ko train kar rahe hai
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))

])
```

```python

#LAST STEP : humne train_test_split karke apne x y ko training and test data me cut kiya(x_test, y_test) aur phir pipeline.fit ko call karke, model ko train kr liya
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
pipeline.fit(x_train, y_train)
```

```python
#predict the value
y_pred = pipeline.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
#thus yeh jo model upar banaa hai, isko hume DOWNLOAD kar lena hai using pickle
import pickle

pickle_model_path = "modelOur.pkl"
with open(pickle_model_path, "wb") as f:
    pickle.dump(pipeline, f)

```

***Now, we will create an END POINT for this model of ours using FAST API***

```python
#here in building API the HTTP method we will use is POST,
#kyu POST? :-> POST is used when a client wants to send some data to a server or client chahta hai ki SERVER is DATA ko process kare and process karke bohot chize ki ja sakta hai
#like ek nayaa resource create ki ja sakti hai, ya phir ek ML Model ka inference karaaya ja sakta hai etc etc
#hum client ki side se REQ BODY me kuch i/p bhejenge..kya i/p bhjna hai? hum kisi ek given user ka sara ka sara i/p features bhejenge like age, wei8, height(ORIGINAL DATA)
#all these features will be validated,once validated then using these features we will calculate the additional features like(bmi, age_grp, lifestyle_risk, cuty tier etcetc)
#and phir yeh saare features ko ML MODEL me bhejenge and model predict karke JSON ke form me o/p dega(HIGH, MEDIUM, LOW) which will be displayed in our front-end
```

```python
from fastapi import FastAPI
from fastapi.responses import JSONResponse
from pydantic import BaseModel, Field, computed_field
from typing import Literal, Annotated
import pickle
import pandas as pd

#we have pasted model.pkl file here which we downloaded from colab

# IMPORTING the ML Model, Using FILE-HANDLING
with open('model.pkl', 'rb') as f: #we wanna open it in read binary mode b/c it's a binary file(rb)
    model = pickle.load(f)

#now we will make an obj of FastAPI
app = FastAPI()

#we will now make a pydantic model to validate incoming data(ORIGINAL DATA)
class UserInput(BaseModel):
    Age: Annotated[int, Field(..., gt=0, lt=120, description='Age of the user')]
    Weight: Annotated[float, Field(..., gt=0, lt=180, description='Weight of the user')]
    Height: Annotated[float, Field(..., gt=0, lt=2.5, description='Height of the user')]
    income_lpa: Annotated[float, Field(..., gt=0, description='Annual Salary of the user in L.P.A')]
    smoker: Annotated[bool, Field(..., description='is user a smoker?')]
    city: Annotated[str, Field(..., description='The City of the user')]
    Occupation: Annotated[Literal['Retired', 'unemployed', 'business', 'Govt_job'], Field(..., description='The City of the user')]  #this will be a Literal kyuki hum isme options denaa chahte hai

#Now hume in sab features se naye features banaane hai like hume (age_grp, bmi, etc etc lifestyle_risk) etc ye sab banaani hai
#Thus we will use computed_fields to compute these values
@computed_field    #for BMI
@property
def bmi(self) -> float:
    return self.weight/(self.height**2)

@computed_field   #FOR lIFESTYLE_RISK
@property
def lifestyle_risk(self) -> str:
    if self.smoker and self.bmi > 30:
        return "high"
    if self.smoker and self.bmi > 27:
        return "medium"
    else:
        return "low"

@computed_field    #for AGE_GROUP
@property
def age_group(self) -> str:
    if self.Age < 25:
        return "young"
    elif self.Age < 45:
        return "adult"
    elif self.Age < 60:
        return "middle_aged"
    return "senior"


tier_1_cities = {"Mumbai", "BLR", "Pondicherry", "Pune", "Chennai", "Kolkata"}
tier_2_cities = {"Jaipur", "Chandigarh", "Indore", "Patna", "Lucknow", "Coimbatore", "Ranchi", "Amristar", "Varanasi", "Guwahati", "Allahabad", "Aligarh", "Gaya"}
@computed_field    #for CITY_TIER
@property
def city_tier(self) -> int:
    if self.city in tier_1_cities:
        return 1
    elif self.city in tier_2_cities:
        return 2
    else:
        return 3

#PYDANTIC MODEL ban gayaaa, now we will make our END POINT(/predict)
@app.post('/predict')
def predict_premium(data: UserInput): #is func ko i/p me user ka data milega and yeh user ka data kaun se type ka hogaa? UserInput type ka hogaa, the class we made above, means REQ BODY se data ayega jo seedha chala jaayega humare pyd model(UserInput) ke paas, pyd model us data ke upar kaam karega , validate karega, computed fields nikaalega aur phir wo palat ke hume (data) ke form me yaha par mil jayega
    #ab hume bas proper i/p format create karna hai, thus ek ek karke row ka data pass karne hain hume humaare model me(TRANSFORMED_DATA JO hai like bmi, age grp etc etc)
    #and ye i/ps jo bheja jayega , ye pandas dataframe ke format me bheja jayega b/c Our RandomForestClassifier is trained upon a dataframe object
    input_df = pd.DataFrame([{   #naya d.f bana raha hu and is dataframe me sirf ek row rakhne wale hai jiske andar we will pass a dict jisme bmi ki value hogi, age_group ki value bhi rahegi and other transformed_features
        'bmi' : data.bmi,
        'Age_group' : data.age_group,
        'Lifestyle_risk' : data.lifestyle_risk,
        'City_tier' : data.city_tier,
        'income_lpa' : data.income_lpa,
        'Occupation' : data.Occupation
    }]) #and finally we stored this d.f in a variable input_df
      #thus yeh ban gayaa humara i/p, jo hum apne ML MOdel ke paas bhejenge

    #upar jo MODEL ko humne import kiya hai, uska PREDICT Func ko hum call karenge and iske andar hum pass kar denge apnaa input_df
    prediction = model.predict(input_df)[0] #issey hume ek list me o/p milega aur us list ka 0th item hume chaiye hogaa and this will be our prediction
     #isi prediction ko hume JSON Format me return karna hai

    return JSONResponse(status_code=200, content={'predicted_category': prediction})



```
